```python
from tkinter import *
import tkinter.font as font
import glob
import numpy as np
import matplotlib.pyplot as plt
import scipy.signal as signal
import scipy.integrate as integrate
from matplotlib.backends.backend_tkagg import FigureCanvasTkAgg, NavigationToolbar2Tk
from matplotlib.ticker import MultipleLocator, AutoMinorLocator
```

```python
figs = []
figs_names = []
j = 17
for line in glob.glob(
        "test/*.txt")[0:10]:  #00_old_data  [997:]    [j*100:j*100+100]
    daco = np.loadtxt(line)
    figs_names.append(line.split("\\")[1])
    sos = signal.butter(10, 20000, 'hp', fs=200_000_000, output='sos',
                        analog=False)  #‘ba’, ‘zpk’, sos
    filtered = signal.sosfilt(sos, daco)

    fig, axs = plt.subplots(figsize=(20, 10))
    axs.plot(np.arange(0, 1, 0.000005),
             np.cumsum(filtered) / 2000,
             linewidth=1,
             color="black")
    plt.locator_params(axis='y', nbins=10)
    plt.locator_params(axis='x', nbins=40)
    axs.xaxis.set_minor_locator(AutoMinorLocator(10))
    axs.yaxis.set_minor_locator(AutoMinorLocator(10))
    axs.tick_params(which='major', length=8, direction="inout")
    axs.tick_params(which='minor', length=4, direction="in")
    axs.title.set_text(line.split("\\")[1])
    axs.set_xlabel('Time [ms]')
    axs.set_ylabel('Intensity [nT]')
    ########################################################ZMAZ
    #     plt.title(line)
    #     print(line)
    #     fig.savefig("trigger\\"+line.split("\\")[1][:-4]+'.png', facecolor="white", edgecolor="white")
    ########################################################ZMAZ
    figs.append(fig)
    plt.close()
#     plt.show()
    del fig
    del axs
```

```python
class MainWindow():
    def callback(self):
        self.current_image_number += 1
        if self.current_image_number == len(self.figs):
            self.master.destroy()

        self.canvas.get_tk_widget().pack_forget()
        self.canvas = FigureCanvasTkAgg(self.figs[self.current_image_number],
                                        master=self.master)
        self.canvas.draw()
        self.canvas.get_tk_widget().pack(side=BOTTOM)

    def callback_breakdown_pulse_cg(self):
        file_object = open('labeled_list.txt', 'a')
        file_object.write(figs_names[self.current_image_number] +
                          '\t breakdown_pulse_cg \n')
        file_object.close()

        self.callback()

    def callback_breakdown_pulse_ic(self):
        file_object = open('labeled_list.txt', 'a')
        file_object.write(figs_names[self.current_image_number] +
                          '\t breakdown_pulse_ic \n')
        file_object.close()

        self.callback()

    def callback_ic_activity(self):
        file_object = open('labeled_list.txt', 'a')
        file_object.write(figs_names[self.current_image_number] +
                          '\t ic_activity \n')
        file_object.close()

        self.callback()

    def callback_microsecond_pulse(self):
        file_object = open('labeled_list.txt', 'a')
        file_object.write(figs_names[self.current_image_number] +
                          '\t microsecond_pulse \n')
        file_object.close()

        self.callback()

    def callback_narrow_bipolar_event(self):
        file_object = open('labeled_list.txt', 'a')
        file_object.write(figs_names[self.current_image_number] +
                          '\t narrow_bipolar_event \n')
        file_object.close()

        self.callback()

    def callback_nothing(self):
        file_object = open('labeled_list.txt', 'a')
        file_object.write(figs_names[self.current_image_number] +
                          '\t nothing \n')
        file_object.close()

        self.callback()

    def callback_return_stroke(self):
        file_object = open('labeled_list.txt', 'a')
        file_object.write(figs_names[self.current_image_number] +
                          '\t return_stroke \n')
        file_object.close()

        self.callback()

    def __init__(self, input_figs, figs_names):

        file_object = open('labeled_list.txt', 'w')
        file_object.close()

        self.master = Tk()
        self.master.title('Oznacovac')
        self.master.geometry("1000x1000")
        self.master.state('zoomed')

        self.button_breakdown_pulse_cg = Button(
            self.master,
            text="breakdown_pulse_cg",
            command=self.callback_breakdown_pulse_cg,
            bg='#0CE84C',
            relief=SOLID,
            borderwidth=4,
            fg='#000000',
            activebackground='#6BE899',
            activeforeground='#aaffaa',
            font=font.Font(family='Helvitica', size=12),
            bd=0)
        self.button_breakdown_pulse_cg.pack(side=TOP, fill=BOTH)

        self.button_breakdown_pulse_ic = Button(
            self.master,
            text="breakdown_pulse_ic",
            command=self.callback_breakdown_pulse_ic,
            bg='#0C60F2',
            relief=SOLID,
            borderwidth=4,
            fg='#000000',
            activebackground='#6F95F2',
            activeforeground='#aaffaa',
            font=font.Font(family='Helvitica', size=12),
            bd=0)
        self.button_breakdown_pulse_ic.pack(side=TOP, fill=BOTH)

        self.button_ic_activity = Button(self.master,
                                         text="ic_activity",
                                         command=self.callback_ic_activity,
                                         bg='#D400DB',
                                         relief=SOLID,
                                         borderwidth=4,
                                         fg='#000000',
                                         activebackground='#DB6FD2',
                                         activeforeground='#aaffaa',
                                         font=font.Font(family='Helvitica',
                                                        size=12),
                                         bd=0)
        self.button_ic_activity.pack(side=TOP, fill=BOTH)

        self.button_microsecond_pulse = Button(
            self.master,
            text="microsecond_pulse",
            command=self.callback_microsecond_pulse,
            bg='#F26A0C',
            relief=SOLID,
            borderwidth=4,
            fg='#000000',
            activebackground='#F2AA6F',
            activeforeground='#aaffaa',
            font=font.Font(family='Helvitica', size=12),
            bd=0)
        self.button_microsecond_pulse.pack(side=TOP, fill=BOTH)

        self.button_narrow_bipolar_event = Button(
            self.master,
            text="narrow_bipolar_event",
            command=self.callback_narrow_bipolar_event,
            bg='#EBE107',
            relief=SOLID,
            borderwidth=4,
            fg='#000000',
            activebackground='#EBE971',
            activeforeground='#aaffaa',
            font=font.Font(family='Helvitica', size=12),
            bd=0)
        self.button_narrow_bipolar_event.pack(side=TOP, fill=BOTH)

        self.button_nothing = Button(self.master,
                                     text="nothing",
                                     command=self.callback_nothing,
                                     bg='#90E80C',
                                     relief=SOLID,
                                     borderwidth=4,
                                     fg='#000000',
                                     activebackground='#AFE872',
                                     activeforeground='#aaffaa',
                                     font=font.Font(family='Helvitica',
                                                    size=12),
                                     bd=0)
        self.button_nothing.pack(side=TOP, fill=BOTH)

        self.button_return_stroke = Button(self.master,
                                           text="return_stroke",
                                           command=self.callback_return_stroke,
                                           bg='#0CF2DD',
                                           relief=SOLID,
                                           borderwidth=4,
                                           fg='#000000',
                                           activebackground='#77F2EF',
                                           activeforeground='#aaffaa',
                                           font=font.Font(family='Helvitica',
                                                          size=12),
                                           bd=0)
        self.button_return_stroke.pack(side=TOP, fill=BOTH)

        self.current_image_number = 0

        self.canvas = Canvas(self.master, width=50, height=20)
        self.canvas.pack(side=BOTTOM)

        self.figs = input_figs

        self.canvas = FigureCanvasTkAgg(self.figs[self.current_image_number],
                                        master=self.master)
        self.canvas.draw()
        self.toolbar = NavigationToolbar2Tk(self.canvas, self.master)
        self.toolbar.update()

        self.canvas.get_tk_widget().pack(side=BOTTOM)

        self.canvas.get_tk_widget().pack(side=BOTTOM)

        self.master.mainloop()


MainWindow(figs, figs_names)
```
